```python
import json

source_trace_filepath = "/mnt/network_drive/lrq/traces/code/sample_0_all_input1_output1_ids_upsample_2.json"

qwen_model_name = "Qwen/Qwen2.5-32B"
with open(source_trace_filepath, "r") as f:
    source_trace = json.load(f)
```

```python
dst_trace = []
dst_trace.append(source_trace[0])  # copy the first metadata entry
for request_data in source_trace[1:]:
    request_data['data']['model'] = qwen_model_name
    dst_trace.append(request_data)

dst_trace_filepath = "/mnt/network_drive/lrq/traces/code/code_qwen.json"    
with open(dst_trace_filepath, "w") as f:
    json.dump(dst_trace, f)
```
