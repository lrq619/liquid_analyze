```python
import re
import matplotlib.pyplot as plt
import numpy as np

# Path to the error log file
run_log_dir = "/mnt/network_drive/lrq/logs/logs_2025-08-31-11-48-15/run_0"
worker_err_file = f"{run_log_dir}/worker_0.err"

# Regex to capture lifespan values
pattern = re.compile(r"lifespan:\s*([0-9.]+)s")

lifespans = []
with open(worker_err_file, "r") as f:
    for line in f:
        match = pattern.search(line)
        if match:
            lifespans.append(float(match.group(1))*1000)

if not lifespans:
    print("No lifespan entries found!")
else:
    # Convert to numpy array
    data = np.array(lifespans)

    # Sort data for CDF
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

    # Plot CDF
    plt.figure(figsize=(6, 5))
    plt.plot(sorted_data, cdf, marker=".", linestyle="none")
    plt.xlabel("Lifespan (ms)")
    plt.ylabel("CDF")
    plt.title("CDF of Memory Object Lifespan")
    plt.grid(True, linestyle="--", alpha=0.6)
    # plt.xlim(0, 200)
    plt.show()
```
