```python

from analyze import analyze_data
from utils import RequestData
import json
log_dirs = [
    "/mnt/network_drive/lrq/logs/logs_2025-05-26-22-27-34/run_0",
    "/mnt/network_drive/lrq/logs/logs_2025-06-24-13-57-20/run_0",
    "/mnt/network_drive/lrq/logs/logs_2025-06-24-15-18-46/run_0",
    "/mnt/network_drive/lrq/logs/logs_2025-06-24-16-38-11/run_0"
]
labels = ["Aggregation", "Mooncake\n(1p3d)", "Mooncake\n(2p2d)", "Mooncake\n(3p1d)"]
memory_metric = "vllm:gpu_cache_usage_perc"
metrics_list = [memory_metric]
requests_list = []
for log_dir in log_dirs:
    report_file_path = analyze_data(log_dir, metrics_list)
    with open(report_file_path, 'r') as f:
        report = json.load(f)

    loadgen_data = report['loadgen_data']
    requests = []
    for request_data in loadgen_data:
        if 'error' not in request_data.keys():
            request = RequestData.init_from_dict(request_data)
            requests.append(request)
            continue
    requests_list.append(requests)
```

```python
import matplotlib.pyplot as plt
import numpy as np
metrics = ["E2E Latency", "TTFT", "TPOT", "Transmission"]
percentile = 95
fig, axes = plt.subplots(nrows=1, ncols=len(metrics), figsize=(16,4))
count = 0
for i, metric in enumerate(metrics):
    p_datas = []
    for requests in requests_list:
        raw_datas = []
        for request in requests:
            if i == 0:
                raw_data = request.ttft + request.waiting_latency + request.decode_latency
            elif i == 1:
                raw_data = request.ttft
            elif i == 2:
                raw_data = request.decode_latency / request.decode_tokens
            elif i == 3:
                raw_data = request.waiting_latency
                if raw_data != 0:
                    count += 1
            raw_datas.append(raw_data)
        if i == 3:
            print(f"There are {count} non-zero waiting latency, there are {len(raw_datas)} total requests")
        p_data = np.percentile(raw_datas, percentile)
        if i == 3:
            print(f"P{percentile} waiting latency: {p_data}")
        p_datas.append(p_data)
    

    axes[i].bar(labels, p_datas)
    axes[i].set_title(f"P{percentile} {metric}")
```
